```python
variables = [getvar(ds, var, 17) for var in ('HGT','LU_INDEX')]
data = xr.merge(variables)
```

```python
from pyproj import Proj, transform
inProj  = Proj("+init=EPSG:2263",preserve_units=True)
outProj = Proj("+init=EPSG:4326") # WGS84 in degrees and not EPSG:3857 in meters)
# swap x,y as mkennedy says
y1,x1 =  (110626.2880, 909126.0155)
print(transform(inProj,outProj,x1,y1))
(-74.2700000001129, 40.46999999990434)
```

```python
wrf_proj = pyproj.Proj(proj='lcc',
                       lat_1=ds.TRUELAT1, lat_2=ds.TRUELAT2,
                       lat_0=ds.MOAD_CEN_LAT, lon_0=ds.STAND_LON,
                       a=6370000, b=6370000)
```

```python
wgs_proj = pyproj.Proj(proj='latlong', datum='WGS84')
e, n = pyproj.transform(wgs_proj, wrf_proj, ds.CEN_LON, ds.CEN_LAT)
```

```python
e,n
```

```python
ds.DX
```

```python
dx, dy = ds.DX, ds.DY
nx, ny = ds.dimensions['west_east'].size, ds.dimensions['south_north'].size

# Lower left corner of the domain
x0 = -(nx-1) / 2. * dx + e
y0 = -(ny-1) / 2. * dy + n

# Get grid values
xx, yy = np.meshgrid(np.arange(nx) * dx + x0, np.arange(ny) * dy + y0)
```

```python
our_lons, our_lats = pyproj.transform(wrf_proj, wgs_proj, xx, yy)
```

```python
ds.XLAT_M
```

```python
ds['DIFF'] = np.sqrt((our_lons - ds.XLONG_M)**2 + (our_lats - ds.XLAT_M)**2)
```

```python
eta_attrs = {attr: ds['ZNU'].getncattr(attr) for attr in ds['ZNU'].ncattrs()}
eta_attrs['axis'] = 'Z'
data['bottom_top'] = xr.DataArray(ds['ZNU'][17], dims='bottom_top', attrs=eta_attrs)
data['south_north'] = xr.DataArray(y, dims='south_north', attrs={'axis': 'Y', 'units': 'degree'})
data['west_east'] = xr.DataArray(x, dims='west_east', attrs={'axis': 'X', 'units': 'degree'})
```

```python
# Define the grid_mapping
data['LambertConformal'] = xr.DataArray(np.array(0), attrs={
    'grid_mapping_name': 'lambert_conformal_conic',
    'earth_radius': 6370000,
    'standard_parallel': (ds.TRUELAT1, ds.TRUELAT2),
    'longitude_of_central_meridian': ds.STAND_LON,
    'latitude_of_projection_origin': ds.MOAD_CEN_LAT
})
for var in data.data_vars:
    data[var].attrs['grid_mapping'] = 'LambertConformal'
    if 'projection' in data[var].attrs:
        del data[var].attrs['projection']
    if 'coordinates' in data[var].attrs:
        del data[var].attrs['coordinates']
        
# Save result
data.to_netcdf('/mnt/e/wrf_tibet/test/wrfout_d02_3.nc')
```
